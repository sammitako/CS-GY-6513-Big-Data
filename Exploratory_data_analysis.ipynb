# File Merge

```python
import pandas as pd
import pyarrow.parquet as pq
from pathlib import Path

# Get a list of all Parquet files in the directory
parquet_files = list(Path('./processed-data-oct/').rglob('*.parquet'))

combined_df = pd.DataFrame()
```

```python
# Loop through each Parquet file (selected columns)
for file in parquet_files:
    table = pq.read_table(file, columns=['name', 'title', 'product', 'description', 'domain', 'country', 'label_product', 'price', 'currency'])
    df = table.to_pandas()
    combined_df = combined_df.append(df, ignore_index=True)
```
